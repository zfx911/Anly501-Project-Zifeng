```python
def search_twitter(query, max_results, tweet_fields, bearer_token=bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}

    url = "https://api.twitter.com/2/tweets/search/all?query={}&max_results={}&{}".format(
        query, max_results, tweet_fields)

    print("--------------", url, "--------------")
    response = requests.request("GET", url, headers=headers)
    # print(response.status_code)
    # print(response.text)

    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
query = "online retailer"
tweet_fields = "tweet.fields=text,author_id,created_at,lang,public_metrics,source"
json_response = search_twitter(query=query,tweet_fields=tweet_fields,bearer_token=bearer_token,max_results=100)
tweets_jsondump=[]
for i in json_response['data']:
    tweets_jsondump.append(i)
with open('online_retailer.json','w') as json_file:
    json.dump(tweets_jsondump,json_file)
```
